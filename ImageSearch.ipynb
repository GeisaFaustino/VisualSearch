## My First Bing Image Search 

This notebook reproduces the tutorial [Call and response: your first Bing Image Search query in Python](https://docs.microsoft.com/en-us/azure/cognitive-services/bing-image-search/quickstarts/python)

** Setting the subscription key to my API key for the Bing API service. **

```python
with open('keys.txt') as f:
    lines = f.readlines()
```

```python
subscription_key = lines[0]
assert subscription_key
```

** Defining the endpoint. **

```python
search_url = "https://api.cognitive.microsoft.com/bing/v7.0/images/search"
```

** Defining the type of image to be looked for. **

```python
search_term = "cats"
```

```python
url_file = "https://twistedsifter.files.wordpress.com/2014/10/chin-fur-makes-cat-look-surprised-banye-china-4.jpg"
```

** Calling out to the Bing search APIs and returning the results as a JSON object. **

```python
import requests

# pass in the API key via the headers dictionary
headers = {"Ocp-Apim-Subscription-Key" : subscription_key}

# pass in the API search term via the params dictionary
params  = {"q": search_term, "license": "public", "imageType": "photo"}
response = requests.get(search_url, headers = headers, params = params)
print(response.url)

response.raise_for_status()

# contains the actual images along with rich metadata such as related items
search_results = response.json()
```

```python
import utils as ut
ut.print_json(search_results)
```

** Extracting the thumbnail URLS for the first 16 results. **

```python
thumbnail_urls = [ img["thumbnailUrl"] for img in search_results["value"][:16] ]
```

** Downloading the thumbnail images and rendering them on a $4 \times 4$ grid. **

```python
%matplotlib inline
import matplotlib.pyplot as plt

from PIL import Image
from io import BytesIO

f, axes = plt.subplots(4, 4, figsize=(32,32))
for i in range(4):
    for j in range(4):
        image_data = requests.get( thumbnail_urls[ i + 4 * j ] )
        image_data.raise_for_status()
        image = Image.open( BytesIO( image_data.content ) )        
        axes[i][j].imshow( image )
        axes[i][j].axis( "off" )
```
